## Hint

You will need to have the two datasets at the same resolution and valid data shape. The data are stored in an image so re-interpolation can be done with the image resizing fanciness available in ... (guess where !)

````python
def imresize (image, shape, *args, **kwargs):
    x = np.linspace(0,1, image.shape[1])
    y = np.linspace (0,1, image.shape [0])
    interp = RegularGridInterpolator((y,x), image)
    
    xnew = np.linspace(0,1, shape[1])
    ynew = np.linspace (0,1, shape [0])
    
    xq, yq = np.meshgrid (xnew,ynew)
    
    znew = interp ((yq,xq))
    return znew
````

````python
# load in ETOPO1 data
etopoH = rasterio.open("../Data/Resources/ETOPO1_Ice_c_geotiff.tif", 'r')
etopoH_img = etopoH.read(1)

# resample
## taken every 4th row and every 4th column
etopoH_img = etopoH_img[::4,::4]

# load in age data
ages = np.load("../Data/Resources/global_age_data.3.6.z.npz")["ageData"]

## resized etopoH to be the same size as 'ages'
etopoH_1 = imresize(etopoH_img, ages.shape, order=1, mode='wrap')
etopoH_1[ np.isnan(ages) ] = np.nan

print (etopoH_img.shape)
print (etopoH_1.shape)
print (ages.shape)
````

````python
from scipy.optimize import curve_fit
````

````python
## Creating a function for fitting of the depth on a scatterplot to a curve. 
def func (x,a,b):
    """Fitting function for the data"""
    
    return (a+b * np.sqrt(x))
````

````python
## Reducing the data to exclude zeros, N/A's and infinites. 
topo_reduced1=topo_reduced[~np.isnan(topo_reduced)]
ages_reduced2=ages_reduced1[~np.isnan(ages_reduced1)]
````

````python
## Plotting this further reduced dataset and fitting a curve using the function created
popt, pcov = curve_fit (func,ages_reduced2,topo_reduced1)
fig = plt.figure(figsize=(15,12), facecolor='none')

xages = np.arange(0,175,1)

plt.scatter(ages_reduced2,topo_reduced1, label='data', s=1, color="Blue")
plt.title ('Depth (m) versus Age (Ma) plot', fontsize = 30)
plt.plot(xages, func(xages,*popt), 'r-', label='fit', linewidth=7)
plt.xlim (-10, 180)
plt.xlabel('Age (Ma)', fontsize =15)
plt.ylabel('Depth (m)', fontsize =15)


pass
````
